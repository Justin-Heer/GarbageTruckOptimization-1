```python
import pandas as pd
import glob
```

```python
locations = pd.read_csv('C:/Users/Aidan/OneDrive - Simon Fraser University (1sfu)/Garbage Route Optimization/poco-allzones.csv')
```

```python
counts = locations.groupby('Garbage Zone').count()
```

```python
paths = glob.glob('C:/Users/Aidan/OneDrive - Simon Fraser University (1sfu)/Garbage Route Optimization/routes/*.csv')

dfs = [pd.read_csv(filename,index_col=0) for filename in paths]
```

```python
results = []
secondResults = []
for i in range(0,len(dfs)):
    df = dfs[i]
    p = paths[i]
    numTrucks = int(p[p.find('trucks')-1])
    zone = int(p[p.find('\\')+5])
    if numTrucks == 3:
        continue
    numHouses = counts[counts.index==zone]['Address'].values[0]
    t = round((sum(df.groupby('vehicleID').max('time')['time'])+ 15 * numHouses + 5400*numTrucks)/3600)
    secondResults.append([zone,numTrucks] + list(((df.groupby('vehicleID').max('time')['time']+ 15 * numHouses + 5400*numTrucks)/3600).values))
    if len(secondResults) < 9:
        secondResults += []
    results.append([zone, numTrucks, t])
    
    
   
  
    
```
