```python
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/Aidan/OneDrive - Simon Fraser University (1sfu)/Garbage Route Optimization/routes/zone5_7trucks.csv',index_col=0)
```

```python
startHouses = []
endHouses = []
pairID = []
vehicleID = []
startLongitude = []
startLatitude = []
endLongitude = []
endLatitude = []
for i in range(0,len(df)-1,2):
    startHouses.append(df.iloc[i]['address'])
    endHouses.append(df.iloc[i+1]['address'])
    
    pairID.append(df.iloc[i]['pairID'])
    vehicleID.append(df.iloc[i]['vehicleID'])
    
    startLongitude.append(df.iloc[i]['longitude'])
    startLatitude.append(df.iloc[i]['latitude'])
    endLongitude.append(df.iloc[i+1]['longitude'])
    endLatitude.append(df.iloc[i+1]['latitude'])
```

```python
queries = pd.DataFrame({'startHouses':startHouses,
                        'endHouses':endHouses,
                       'pairID':pairID,
                       'vehicleID':vehicleID,
                       'startLongitude':startLongitude,
                       'startLatitude':startLatitude,
                       'endLongitude':endLongitude,
                       'endLatitude':endLatitude})
```
